### Q1. Write a program to distinguish between Array Indexing and Fancy Indexing.

```python
import numpy as np 

arr = np.array([1,2,3,4,5,6])

# Array Indexing
print("Array Indexing : ")
print("Array[0] : ", arr[0])
print("Array[2] : ", arr[2])
print("Array[5] : ", arr[5])

print()
# Fancy Indexing
print("Array Indexing : ")
print("Array[0], Array[2], Array[5] : ", arr[[0, 2, 5]])
```

### Q2. Execute the 2D array Slicing.

### Q5. Perform the Stack functions in Numpy arrays – Stack(), hstack(), vstack(), and dstack().

```python
import numpy as np

arr1 = np.array([
    [1,2,3],
    [4,5,6]
    ])

arr2 = np.array([
    [7,8,9],
    [4,5,6]
])

print("Stack : ", np.stack((arr1, arr2)))
print()
print("Horizontal Stack : ", np.hstack((arr1,arr2)))
print()
print("Vertical Stack :", np.vstack((arr1,arr2)))
print()
print("DStack :", np.dstack((arr1,arr2)))
```

### Q6. Perform the searchsort method in Numpy array.

##### numpy.searchsorted(a, v, side='left', sorter=None)
Find indices where elements should be inserted to maintain order.

Find the indices into a sorted array a such that, if the corresponding elements in v were inserted before the indices, the order of a would be preserved.

```python
import numpy as np

arr = np.array([1,2,3,4,8,9,10])
new = np.array([5,11])
print(np.searchsorted(arr, new))
```

### Q9. Create Data frame on your Domain area and perform the following operations to find and eliminate the missing data from the dataset.
• isnull()
• notnull()
• dropna()
• fillna()
• replace()
• interpolate()

```python
import pandas as pd
import numpy as np


df = pd.read_csv("./event_management_dataset.csv")

# Display the DataFrame
print("Original DataFrame:")
print(df)

# Check for missing data using isnull()
print("\nChecking for missing data using isnull():")
print(df.isnull())

# Check for non-missing data using notnull()
print("\nChecking for non-missing data using notnull():")
print(df.notnull())

# Drop rows with missing data using dropna()
print("\nDropping rows with missing data using dropna():")
df_dropna = df.dropna()
print(df_dropna)

# Fill missing data using fillna()
print("\nFilling missing data using fillna():")
df_fillna = df.fillna(0)  # Filling missing data with 0 (you can use other values as well)
print(df_fillna)

# Replace missing data using replace()
print("\nReplacing missing data using replace():")
df_replace = df.replace(np.nan, -1)  # Replace NaN with -1 (you can use other values as well)
print(df_replace)

# Interpolate missing data using interpolate()
print("\nInterpolating missing data using interpolate():")
df_interpolate = df.interpolate()
print(df_interpolate)
```
