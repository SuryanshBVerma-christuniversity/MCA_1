```python
import pandas as pd
df = pd.read_csv("./event_management_dataset.csv")
```

```python
from matplotlib import pyplot as plt

# Median Developer Salaries by Age
x = df["Event ID"][0:10]

y = df["Tickets Sold"][0:10]

plt.plot(x, y)
plt.xlabel('Events')
plt.ylabel('ickets Sold')
plt.title('')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = df["Attendees"][0:10]
y = df["Tickets Sold"][0:10]

# Set the figure size in inches
plt.figure(figsize=(10,6))

# Scatter plot with a label
plt.scatter(x, y, label="Data Points")

# Set x and y axes labels
plt.xlabel('X Values')
plt.ylabel('Y Values')

plt.title('Scatter Title')

# Add legend with the label "Data Points"
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = df["Attendees"][0:10]
y = df["Date"][0:10]
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(y, x)
 
plt.xlabel("Date")
plt.ylabel("Attendees")
plt.title("Attendees Vs Regsitration Fees")
plt.show()
```
